```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
x = [2,4,2.2,3.1,7,4.6]
y = [3,1,5,7.2,8,4.7,]
```

```python
plt.plot(x, y, '.')
```

```python
x1 = np.random.normal(185,10,250)
y1 = np.random.normal(185,10,250)
plt.plot(x1, y1, '.')
#print(x1)
```

```python
## Covariance indicates the direction of the linear relationship between the variables 
# Correlation coeficient(r) on the other hand measures both direction and strenght between the variables
# Positive Correlation: both variables change in the same direction (r <= 1)
# Neutral Correlation: No relationship in the change of the variables (r = 0 | close to 0)
# Negative Correlation: variables change in opposite directions (r >= -1)
```

```python
# Correlation:
# We can calculate correlation between the variable with scipy.stats.pearsonr in python it returns Pearson’s correlation coefficient and p-value
# P-value is the probability of finding the observed (If the p-value is less then selected alpha, we reject the null hypothesis)
from scipy.stats import pearsonr 
corr = pearsonr(x1, y1)
print('Pearson’s correlation coefficient: {:.2f}'.format(corr[0]))
print('P-value: {:.2f}'.format(corr[1]))
```

```python
plt.hist(x1)
print('mean: {:.2f}'.format(np.mean(x1)))
print('median: {:.2f}'.format(np.median(x1)))
print('STD: {:.2f}'.format(np.std(x1)))
m = plt.axvline(np.mean(x1), color='g', label='mean')
med = plt.axvline(np.median(x1), color='r', label='median')
plt.legend(loc='best')
plt.show()
```

```python
plt.hist(y1)
m = plt.axvline(np.mean(y1), color='g', label='mean')
med = plt.axvline(np.median(y1), color='r', label='median')
plt.legend(loc='best')
print('mean: {:.2f}'.format(np.mean(y1)))
print('median: {:.2f}'.format(np.median(y1)))
print('STD: {:.2f}'.format(np.std(y1)))
```

```python
# Z-score is also called as standard score gives you an idea of how far from the mean a data point is. Technically it's a measure of how many standard deviation below or above the population mean.
from scipy.stats import zscore
z_score = zscore(x1, axis=None)
print('Z_Score of fisrt data point: ', z_score[0])
print('mean: {:.2f}'.format(np.mean(x1)))
print('STD: {:.2f}'.format(np.std(x1)))
print('First data point: ', x1[0])
```
